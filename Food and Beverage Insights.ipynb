```python
# Join survey_responses_df with respondents_df based on 'Respondent_ID'
merged_df = survey_responses_df.merge(respondents_df, on='Respondent_ID', how='left')

# Join merged_df with cities_df based on 'City_ID'
final_df = merged_df.merge(cities_df, on='City_ID', how='left')

# Group by 'City' and calculate the average consumption frequency
consumption_freq_by_city = final_df.groupby('City')['Consume_frequency'].count().reset_index()

# Sort the DataFrame by count in descending order
consumption_freq_by_city = consumption_freq_by_city.sort_values('Consume_frequency', ascending=False)

# Plot the city-wise consumption frequency
plt.figure(figsize=(12, 6))
sns.barplot(x='City', y='Consume_frequency', data=consumption_freq_by_city)
plt.xlabel('City')
plt.ylabel('Consumption Frequency')
plt.title('Energy Drink Consumption Frequency by City')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by 'Tier' and 'Marketing_channels' and calculate the count of consumers
marketing_channel_reach = final_df.groupby(['Tier', 'Marketing_channels'])['Respondent_ID'].count().reset_index()

# Pivot the data to create a heatmap
heatmap_data = marketing_channel_reach.pivot(index='Marketing_channels', columns='Tier', values='Respondent_ID')

# Plot the heatmap of marketing channel reach in different city tiers
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='d', cbar=True)
plt.xlabel('City Tier')
plt.ylabel('Marketing Channels')
plt.title('Marketing Channel Reach in Different City Tiers')
plt.show()
```

```python
# Group by 'Taste_experience' and 'Brand_perception' and calculate the count of respondents
taste_vs_brand = final_df.groupby(['Taste_experience', 'Brand_perception'])['Respondent_ID'].count().reset_index()

# Pivot the data to create a stacked bar chart
pivot_data = taste_vs_brand.pivot(index='Taste_experience', columns='Brand_perception', values='Respondent_ID')

# Plot the stacked bar chart of taste experience vs brand perception
plt.figure(figsize=(10, 6))
pivot_data.plot(kind='bar', stacked=True)
plt.xlabel('Taste Experience')
plt.ylabel('Count')
plt.title('Taste Experience vs Brand Perception')
plt.legend(title='Brand Perception')
plt.xticks(rotation=0)
plt.show()
```

```python
# Group by 'Tier' and 'Typical_consumption_situations' and calculate the count of consumers
consumption_situations = final_df.groupby(['Tier', 'Typical_consumption_situations'])['Respondent_ID'].count().reset_index()

# Pivot the data to create a stacked bar chart
pivot_table = consumption_situations.pivot(index='Tier', columns='Typical_consumption_situations', values='Respondent_ID')

# Plot the stacked bar chart of consumption situations in different city tiers
plt.figure(figsize=(10, 6))
pivot_table.plot(kind='bar', stacked=True)
plt.xlabel('City Tier')
plt.ylabel('Count')
plt.title('Typical Consumption Situations among Energy Drink Consumers in Different City Tiers')
plt.legend(title='Consumption Situations')
plt.show()
```

```python
# Group by 'Tier' and 'Consume_reason' and calculate the count of respondents
reasons_by_tier = final_df.groupby(['Tier', 'Consume_reason'])['Respondent_ID'].count().reset_index()

# Identify the most common reasons for consuming energy drinks in each city tier
top_reasons_by_tier = reasons_by_tier.groupby('Tier').apply(lambda x: x.nlargest(3, 'Respondent_ID')).reset_index(drop=True)

# Plot a grouped bar chart to visualize the most common reasons for consuming energy drinks in each city tier
plt.figure(figsize=(12, 6))
sns.barplot(x='Consume_reason', y='Respondent_ID', hue='Tier', data=top_reasons_by_tier)
plt.xlabel('Consumption Reason')
plt.ylabel('Count')
plt.title('Most Common Reasons for Consuming Energy Drinks by City Tier')
plt.xticks(rotation=45)
plt.legend(title='City Tier')
plt.show()
```
